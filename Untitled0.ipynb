```python
#fucntionality used
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import urllib.request, json
import os
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
import yfinance as yf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
```

```python
data = yf.download("RELIANCE.NS", start="2012-01-01", end="2025-01-01", interval="1d")

print(data.head())
```

```python
# Sort DataFrame by date
data = data.sort_values('Date')
print(data.shape[0])

# Double check the result
data.head()
```

```python
plt.figure(figsize=(18,9))
plt.plot(data.index, (data['Low'] + data['High']) / 2.0)
plt.xticks(data.index[::500], rotation=45)

plt.xlabel('Date', fontsize=18)
plt.ylabel('Mid Price', fontsize=18)
plt.show()
```

```python
# Cell 2: Prepare your data (assuming 'data' DataFrame with 'Date' and 'Mid Price')

high_prices = data['High'].values
low_prices = data['Low'].values
mid_prices = (high_prices + low_prices) / 2.0

# Convert mid_prices to a DataFrame for easier processing
mid_prices = pd.DataFrame(mid_prices, columns=['Mid_Price'])
```

```python
# Cell 3: Normalize the data
scaler = MinMaxScaler(feature_range=(0,1))

scaled_data = scaler.fit_transform(mid_prices)
```

```python
# Cell 4: Create sequences for LSTM
def create_sequences(data, window_size=60):
    X = []
    y = []
    for i in range(window_size, len(data)):
        X.append(data[i-window_size:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

window_size = 60
X, y = create_sequences(scaled_data, window_size)

# Reshape X to be 3D for LSTM: (samples, timesteps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# Cell 5: Build the LSTM model
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

```python
# Cell 6: Train the model
history = model.fit(X, y, epochs=50, batch_size=32, validation_split=0.1)
```

```python
# Cell 7: Plot training & validation loss
plt.figure(figsize=(10,6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Cell 8: Prepare test data (use last part of your data as test set)
test_size = 200
total_data = np.concatenate((scaled_data,))

# Prepare test sequences
X_test = []
y_test = []

for i in range(len(scaled_data) - test_size, len(scaled_data)):
    X_test.append(scaled_data[i-window_size:i, 0])
    y_test.append(scaled_data[i, 0])

X_test = np.array(X_test)
y_test = np.array(y_test)
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
# Cell 9: Make predictions on test data
predicted_mid_prices = model.predict(X_test)
predicted_mid_prices = scaler.inverse_transform(predicted_mid_prices)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
# Cell 10: Visualize predictions vs actual mid prices
plt.figure(figsize=(16,8))
plt.plot(data.reset_index()['Date'][-test_size:], y_test_actual, color='blue', label='Actual Mid Price')
plt.plot(data.reset_index()['Date'][-test_size:], predicted_mid_prices, color='red', label='Predicted Mid Price')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Mid Price', fontsize=14)
plt.title('LSTM Predicted vs Actual Mid Price', fontsize=18)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
